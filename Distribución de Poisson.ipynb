Caso de negocio:

En el emocionante mundo de las estadísticas y la probabilidad, la Distribución de Poisson desempeña un papel esencial. Esta distribución se utiliza para modelar la probabilidad de que ocurra un número específico de eventos en un intervalo de tiempo dado. Pero, ¿cómo se relaciona esto con el negocio de seguros? Hoy, exploraremos el poder de la Distribución de Poisson y su aplicación en la predicción de reclamosen la industria del seguro.

La Distribución de Poisson en un Vistazo:

La Distribución de Poisson es especialmente útil cuando tratamos con eventos que ocurren de manera aleatoria e independiente, y cuando conocemos la tasa promedio de ocurrencia (llamada "media"). Imaginemos que tenemos una empresa de seguros con un promedio de 10 reclamaciones por día. Aquí, la media es de 10 reclamos.

Calculando la Probabilidad:

¿Qué sucede si queremos calcular la probabilidad de que exactamente 12 reclamos ocurran en un día? Utilizamos la fórmula de la Distribución de Poisson:


P(X = k) = (e^(-μ) * μ^k) / k!
P(X = k) es la probabilidad de que ocurran k eventos.
e es la constante matemática (~2.71828).
μ es la media (en nuestro caso, 10 reclamos por día).
k es el número deseado de eventos (en este caso, 12 reclamos ).
k! es el factorial de k.
Sustituyendo los valores, calculamos P(X = 12) y encontramos que la probabilidad es aproximadamente 0.0948, lo que significa que hay alrededor del 9.48% de probabilidad de que ocurran exactamente 12 reclamos en un día.

```python
import math 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definimos la media de la distribución de Poisson
mu = 10
```

```python
# Calculamos la probabilidad de que haya 12 reclamos
p = np.exp(-mu) * mu ** 12 / math.factorial(12)
```

```python
# Creamos un nuevo array de índices enteros
x = np.arange(0, 20)
```

```python
# Convertimos el array de flotantes a un array de enteros escalares
x = np.asarray(x, dtype=int)
```

```python
# Calculamos la probabilidad de Poisson para cada valor en x
poisson_pmf = np.exp(-mu) * mu ** x / [math.factorial(val) for val in x]
```

```python
# Graficamos la distribución de probabilidad
plt.plot(x, poisson_pmf, marker='o', linestyle='-', label='Distribución de Poisson')
plt.fill_between(x, 0, poisson_pmf, where=(x <= 12), color='gray', alpha=0.5)
plt.title("Distribución de Poisson")
plt.xlabel("Número de reclamaciones")
plt.ylabel("Probabilidad")
plt.text(12, 0.03, f'P(12) = {p*100:.2f}%', fontsize=12, ha='right', va='bottom')

plt.grid(True)
plt.legend()
```
